# Neural Networks for Handwritten Digits — Short Summary

This notebook summarizes the key concepts from the Neural Networks practice lab.


# 1. Dataset

- Images are `20 × 20`
- Flattened into `400` numbers

Formula:

$$20 \times 20 = 400$$

So:

```python
X.shape = (1000, 400)
y.shape = (1000, 1)
```

- `X` = images
- `y` = labels (0 or 1)


# 2. Neural Network Structure

```text
400 → 25 → 15 → 1
```

- 400 inputs
- Hidden layer 1 → 25 neurons
- Hidden layer 2 → 15 neurons
- Output → 1 neuron


# 3. Core Formula

Every neuron does:

$$a = g(Wx+b)$$

Where:
- `x` = input
- `W` = weights
- `b` = bias
- `g` = activation function
- `a` = output


````
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
````

# 4. Building Model in TensorFlow

````
model = Sequential([
    tf.keras.Input(shape=(400,)),
    Dense(25, activation='sigmoid'),
    Dense(15, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])

model.summary()
````

# 5. Train Model

````
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.001)
)

# Train model
# model.fit(X, y, epochs=20)
````

# 7. Manual Dense Layer (Important)

This manually performs:
1. Dot product
2. Add bias
3. Apply activation


````
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def my_dense(a_in, W, b, g):

    units = W.shape[1]
    a_out = np.zeros(units)

    for j in range(units):

        w = W[:, j]

        z = np.dot(w, a_in) + b[j]

        a_out[j] = g(z)

    return a_out
````

# 8. Vectorized Version

Fast matrix implementation using NumPy.


````
def my_dense_v(A_in, W, b, g):

    Z = np.matmul(A_in, W) + b

    A_out = g(Z)

    return A_out
````

# 9. Important Matrix Rule

$$(m \times n)(n \times j)=(m \times j)$$
